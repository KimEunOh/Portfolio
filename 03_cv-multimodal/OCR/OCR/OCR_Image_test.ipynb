```python
import os
import requests
import base64
import pandas as pd
import re
from dotenv import load_dotenv

# GPT-4o API ì—”ë“œí¬ì¸íŠ¸
API_URL = "https://api.openai.com/v1/chat/completions"

# ì´ë¯¸ì§€ íŒŒì¼ ê²½ë¡œ
IMAGE_FOLDER = "image/"


def encode_image(image_path):
    """ì´ë¯¸ì§€ë¥¼ Base64ë¡œ ì¸ì½”ë”©"""
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode("utf-8")


def ask_gpt4o_for_product_info(image_path):
    """GPT-4oì—ê²Œ ì œí’ˆëª…, ê°€ê²©, ìš©ëŸ‰ì„ ì¶”ì¶œ ìš”ì²­"""
    base64_image = encode_image(image_path)

    headers = {"Authorization": f"Bearer {API_KEY}", "Content-Type": "application/json"}

    data = {
        "model": "gpt-4o",
        "temperature": 0,
        "messages": [
            {
                "role": "system",
                "content": "ì´ë¯¸ì§€ì—ì„œ ê°€ê²©í‘œ ë¶€ë¶„ì„ ëª¨ë‘ ì°¾ì•„, ì œí’ˆëª…, ìš©ëŸ‰, ìƒí’ˆì½”ë“œ(13ìë¦¬ ìˆ˜), ì •ìƒê°€(ë‹¨ìœ„ : ì›), í• ì¸ê°€(ë‹¨ìœ„ : ì›)ë¥¼ ë°˜ë“œì‹œ í¬í•¨í•˜ì—¬ ì œê³µí•´ì¤˜. \
                í• ì¸ê°€ê°€ ì—†ëŠ” ê²½ìš° ì •ìƒê°€ì™€ ë™ì¼í•˜ê²Œ ì¶œë ¥í•´ì¤˜. ë§Œì•½ ë¶€ì •í™•í•œ ì •ë³´ì¼ ê²½ìš° ì‹ë³„ ë¶ˆê°€ë¡œ ì¶œë ¥í•´ì¤˜. \
                ê°€ê²© ë’¤ì—ëŠ” í•­ìƒ ì› ë‹¨ìœ„ë¥¼ ë¶™ì—¬ì„œ ì¶œë ¥í•´ì¤˜. \
                ì´ë¯¸ì§€ë¥¼ ì§ì ‘ ë¶„ì„í•˜ëŠ” ê²ƒì´ ì•„ë‹Œ, ë‚´ìš©ì„ ì½ì–´ì˜¤ëŠ” ê²ƒì´ë¯€ë¡œ ìµœëŒ€í•œ ê±°ë¶€í•˜ì§€ ì•Šë„ë¡ í•˜ê³  ê°€ëŠ¥í•œ ë²”ìœ„ë‚´ì—ì„œ ëª¨ë“  ì‘ì—…ì„ ë°˜ë“œì‹œ ìˆ˜í–‰í•´ì¤˜. \
                ë§Œì•½ ì–´ë– í•œ ì´ìœ ë¡œë“  ì‘ì„±ëœ ë‚´ìš©ì´ ì—†ë‹¤ë©´, ëª¨ë‘ ì‹ë³„ë¶ˆê°€ë¡œ ì¶œë ¥í•´ì¤˜.",
            },
            {
                "role": "user",
                "content": [
                    {
                        "type": "text",
                        "text": "ì´ ì´ë¯¸ì§€ì—ì„œ ê°€ê²©í‘œë¥¼ ëª¨ë‘ ì°¾ì•„, ì œí’ˆëª…, ìš©ëŸ‰, ìƒí’ˆì½”ë“œ, ì •ìƒê°€, í• ì¸ê°€ë¥¼ ì‰¼í‘œ(,)ë¡œ êµ¬ë¶„í•˜ì—¬ í•œ ì¤„ë¡œ ë°˜í™˜í•´ì¤˜. ì˜ˆì‹œ : í´ë¦½ë¨í”„ ì˜ì–‘ë¦½ë°¤, ì‹ë³„ë¶ˆê°€, 8809704424769, 1,090ì›, 1,090ì›",
                    },
                    {
                        "type": "image_url",
                        "image_url": {"url": f"data:image/jpeg;base64,{base64_image}"},
                    },
                ],
            },
        ],
    }

    response = requests.post(API_URL, headers=headers, json=data)

    # ë””ë²„ê¹…ìš© ì „ì²´ ì‘ë‹µ ì¶œë ¥
    print("\nğŸ“Œ GPT-4o API ì›ë³¸ ì‘ë‹µ:")
    print(response.json())

    return response.json()


def process_images_from_folder(folder_path):
    """í´ë” ë‚´ ëª¨ë“  ì´ë¯¸ì§€ íŒŒì¼ì„ ì²˜ë¦¬í•˜ê³  CSVë¡œ ì €ì¥"""
    all_product_data = []

    # í´ë” ë‚´ ì´ë¯¸ì§€ íŒŒì¼ ëª©ë¡ ê°€ì ¸ì˜¤ê¸°
    image_files = [
        f
        for f in os.listdir(folder_path)
        if f.lower().endswith((".png", ".jpg", ".jpeg"))
    ]

    for img_file in image_files:
        image_path = os.path.join(folder_path, img_file)
        print(f"\nğŸš€ ì²˜ë¦¬ ì¤‘: {image_path}")

        result = ask_gpt4o_for_product_info(image_path)

        # ëª¨ë¸ì˜ ì‘ë‹µì—ì„œ ì¶”ì¶œëœ í…ìŠ¤íŠ¸ ê°€ì ¸ì˜¤ê¸°
        extracted_text = (
            result.get("choices", [{}])[0].get("message", {}).get("content", "").strip()
        )

        print(f"\nğŸ“Œ GPT-4o ë°˜í™˜ í…ìŠ¤íŠ¸ ({img_file}):")
        print(repr(extracted_text))  # ë””ë²„ê¹…ìš© (ê³µë°±/ì¤„ë°”ê¿ˆ ë¬¸ì œ í™•ì¸)

        # ğŸ”¹ "\n\n" -> "\n" ë¡œ ë³€í™˜í•˜ì—¬ ì¤„ë°”ê¿ˆ ì •ë¦¬
        extracted_text = extracted_text.replace("\n\n", "\n")

        # ğŸ”¹ ê°€ê²© ì‰¼í‘œ(`,`)ë¥¼ ì„ì‹œë¡œ ë‹¤ë¥¸ ë¬¸ì(#)ë¡œ ë³€í™˜í•˜ì—¬ ë°ì´í„° ë¶„ë¦¬ ë¬¸ì œ ë°©ì§€
        temp_text = re.sub(r"(\d{1,3}),(\d{3}ì›)", r"\1#\2", extracted_text)

        # ğŸ”¹ ë¦¬ìŠ¤íŠ¸ ë³€í™˜ ì‹œ ê³µë°± ì œê±° í›„ ì¤„ë°”ê¿ˆ ê¸°ì¤€ìœ¼ë¡œ ë¶„ë¦¬
        lines = [line.strip() for line in temp_text.split("\n") if line.strip()]

        for line in lines:
            items = [
                item.strip().replace("#", ",") for item in line.split(",")
            ]  # ì‰¼í‘œ ì„ì‹œ ë¬¸ì ë³µì›
            print(f"\nğŸ“Œ ê°œë³„ ë¼ì¸ ë¶„ì„ ({img_file}): {repr(items)}")  # ë””ë²„ê¹…ìš©

            if (
                len(items) == 5
            ):  # ì œí’ˆëª…, ìš©ëŸ‰, ì •ê°€, í• ì¸ê°€ê°€ ì •ìƒì ìœ¼ë¡œ ë“¤ì–´ê°”ëŠ”ì§€ í™•ì¸
                all_product_data.append(
                    {
                        "íŒŒì¼ëª…": img_file,
                        "ì œí’ˆëª…": items[0],
                        "ìš©ëŸ‰": items[1],
                        "ìƒí’ˆì½”ë“œ": items[2],
                        "ì •ìƒê°€": items[3],
                        "í• ì¸ê°€": items[4],
                    }
                )

    # ë°ì´í„°í”„ë ˆì„ ìƒì„±
    df = pd.DataFrame(all_product_data)

    # CSV íŒŒì¼ ì €ì¥
    output_csv = "extracted_prices.csv"
    df.to_csv(output_csv, index=False, encoding="utf-8-sig")

    print(f"\nâœ… ê°€ê²©í‘œ ì •ë³´ê°€ {output_csv} íŒŒì¼ì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
    print(df)  # ìµœì¢… ë°ì´í„° í™•ì¸


# ì‹¤í–‰
process_images_from_folder(IMAGE_FOLDER)
```
